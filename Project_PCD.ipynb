```python
# untuk menampilkan citra hasil olahan
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#untuk melakukan operasi numerik
import numpy as np
import scipy
#untuk memuat citra sebagai matriks
from scipy import ndimage
# fungsi Image didapatkan dalam folder
from scipy.misc.pilutil import Image
%matplotlib inline
```

```python
#untuk mengimpor gambar dari sebuah folder
gambar = "D:\Random\pantai.jpg"
img = mpimg.imread(gambar)
plt.imshow(img)
plt.title('Citra yang digunakan')
```

```python
#Digunakan untuk menampilkan pada citra awal

def plot(data, title):
    plot.i += 1
    plt.subplot(2, 2, plot.i)
    plt.imshow(data)
    plt.gray()
    plt.title(title)
    
plot.i = 0
im = Image.open(gambar).convert('L
# untuk menjadikan citra yang sebelumnya disimpan dalam variabe im
data1 = np.array(im, dtype=int)
# untuk memplotkan matrik 2 dimensi dari variabel data dengan title citra masukan.
plot(data1, 'Gambar Asli')

# perintah untuk membuat filter yang dituliskan berformat matrix
kernel_awal = np.array([[-1, -1, -1],
                  [-1, 8, -1],
                  [-1, -1, -1]])
# melakukan operasi konvolusi terhadap citra asli yang dituliskan dengan variabel data untuk menampilkan hasil keseluruhan
highpass_3x3 = ndimage.convolve(data1, kernel_awal)
plot(highpass_3x3, 'High Pass 3x3')
#  untuk menampilkan hasil olahan secara kelesuruhan menggunakan fungsi plt
plt.show()

hist, bin_edges = scipy.histogram(data1, bins=range(300))
plt.bar(bin_edges[:-1], hist, width = 1)
plt.xlim(min(bin_edges), max(bin_edges))
plt.title('Histogram Gambar Asli')
plt.show()

hist, bin_edges = scipy.histogram(highpass_3x3, bins=range(300))
plt.bar(bin_edges[:-1], hist, width = 1)
plt.xlim(min(bin_edges), max(bin_edges))
plt.title('Histogram Highpass 3x3')
plt.show()
```

```python
# fungsi yang bernama plot dungan deklarasi pada python menggunakan perintah def
def plot(data, title
# proses increment pada variabel i
    plot.i += 1
# untuk melakukan subplotting atau melakukan plotting citra lain
    plt.subplot(2,2,plot.i)
#  menampilkan citra asli yang dituliskan dengan variabel data
    plt.imshow(data)
# perintah untuk menampilkan citra dalam grayscale
    plt.gray
#  perintah untuk menampilkan title citra masing masing olahan
    plt.title(title
#  nilai i untuk inisialisasi adalah im = Image.open
plot.i = 0

#fungsi convert untuk meneruskan format warna yang diinginkan
im = Image.open(gambar).convert('L')
data3 = np.array(im, dtype=float)
plot(data3, 'Gambar Asli')

#perhitungan manual
np.set_printoptions(threshold=np.inf)
print(data3, "Matriks Citra Asli")

#kernel 2x2 piksel untuk menghasilkan pada tepi atas atau tepi bawah
kernel_robert_1 = np.array([[1, 0],
                          [0, -1]
                         ])
# melakukan operasi konvolusi terhadap citra asli 
hrobert_2x2 = ndimage.convolve(data3, kernel_robert_1)
plot(hrobert_2x2, 'High Pass 2x2 Robert Horizontal')
plt.show()

kernel_robert_2 = np.array([[0, 1],
                            [-1,0]])
vrobert_2x2 = ndimage.convolve(data3, kernel_robert_2)
plot(vrobert_2x2, 'High Pass 2x2 Robert Vertical')
plt.show()

hvrobert_2x2 = np.sqrt(abs(hrobert_2x2)**2 + abs(vrobert_2x2)**2)
plot(hvrobert_2x2, 'Hasil Akhir Robert')
plt.show()

#perhitungan manual
np.set_printoptions(threshold=np.inf)
print(hvrobert_2x2, "Matriks Robert")
#print(hvrobert_2x2)

hist, bin_edges = scipy.histogram(hrobert_2x2, bins=range(45))
plt.bar(bin_edges[:-1], hist, width = 1)
plt.xlim(min(bin_edges), max(bin_edges))
plt.title('Histogram Robert Horizontal 2x2')
plt.show()

hist, bin_edges = scipy.histogram(vrobert_2x2, bins=range(45))
plt.bar(bin_edges[:-1], hist, width = 1)
plt.xlim(min(bin_edges), max(bin_edges))
plt.title('Histogram Robert Vertical 2x2')
plt.show()

hist, bin_edges = scipy.histogram(hvrobert_2x2, bins=range(45))
plt.bar(bin_edges[:-1], hist, width = 1)
plt.xlim(min(bin_edges), max(bin_edges))
plt.title('Histogram Robert')
plt.show()
```
